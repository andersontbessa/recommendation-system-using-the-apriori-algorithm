```python
df  = pd.read_excel('campanhas.xlsx')

#dados.columns[7]
#dados.isnull().sum()
df.rename(columns={'Das tags citadas abaixo, com quais você se identifica e provavelmente participaria de uma ação envolvendo essas palavras? (Selecione quantas quiser)'
:'tag'}, inplace=True)
df.rename(str.lower, axis='columns')
tag=df['tag'].str.split(';', expand=True)


pd.set_option('display.max_columns', None)
```

```python
tag.replace(to_replace=[None], value=0, inplace=True)
tag.head(5)

#tag.shape
```

```python
def removeAllOcurrencesOfValueInList(_list, value):
    return list(filter(lambda x: x != value, _list))
```

```python

list_all_transactions = []

for index, row in tag.iterrows():
    list_single_transaction = row.values.tolist()
    list_single_transaction = removeAllOcurrencesOfValueInList(list_single_transaction, 0)
    list_all_transactions.append(list_single_transaction)

print(list_all_transactions[0:10])
```

```python

from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
#http://rasbt.github.io/mlxtend/user_guide/preprocessing/TransactionEncoder/
te_ary = te.fit(list_all_transactions).transform(list_all_transactions)
trans = pd.DataFrame(te_ary, columns=te.columns_)

pd.set_option('display.max_columns', None)
trans.head()
```

```python
trans.rename(columns={'Cult. Regional (voluntariado de pessoas da dell de diferentes áreas do brasil fazendo voluntariado para diferentes organizações de diferentes lugares, para criar mais empatia entre team members a nível nacional)':'Cult. Regional'}, inplace=True)

#trans['@maisqueobasico atende famílias indígenas também, começamos com eles.'].value_counts().head()
trans.drop('@maisqueobasico atende famílias indígenas também, começamos com eles.', axis=1, inplace=True)
#trans.drop('Cult. Regional (voluntariado de pessoas da dell de diferentes áreas do brasil fazendo voluntariado para diferentes organizações de diferentes lugares, para criar mais empatia entre team members a nível nacional)', axis=1, inplace=True)
trans.drop('All volunteer works are valid.', axis=1, inplace=True)


#trans.head()
pd.set_option('display.max_columns', None)
trans.head(6)
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

frequent_itemsets = apriori(trans, min_support = 0.01, use_colnames = True)
frequent_itemsets.sort_values(by=['support'], ascending = False).head(10)
```

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.3)
rules.sort_values(by=['lift'], ascending = False).drop(['antecedent support', 'consequent support', 'leverage', 'conviction'], axis=1)
```
